```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
```

```python
df=pd.read_csv("C:/Users/Bala Subaranjitha B/Desktop/pdfs/homeprices.csv")
```

```python
df=pd.DataFrame(df)
```

```python
df
```

```python
X=np.array(df['SquareFeet'])
```

```python
Y=np.array(df['Price'])
```

```python
plt.plot(X,Y,marker='o',color='r',ms=10,mec='blue',mfc='blue')
plt.xlabel('Area in Square Feet')
plt.ylabel('Price in Rupees')
plt.show()
```

```python
#Linear Regression
reg=lm.LinearRegression()
reg.fit(df[['SquareFeet']],df.Price)
```

```python
reg.predict([[2300]])
```

```python
plt.plot(X,reg.predict(df[['SquareFeet']]))
plt.plot(X,Y,'o')
plt.show()
```
